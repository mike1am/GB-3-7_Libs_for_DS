```python
%matplotlib widget
```

```python
n, m = 4, 3
A = np.arange(n * m).reshape(n, -1)
At = A.T
B = A @ At
tools.showVectors(B)
B.shape, la.matrix_rank(A), la.matrix_rank(B)
```

Ранг матрицы A равен 2, что меньше n, т.к. она составлена из арифметической последовательности и все её строки, кроме 2-х, линейно зависимы. Ранг произведения не может быть больше. Соответственно, обратная матрица для матрицы B размера nxn не может быть получена.
